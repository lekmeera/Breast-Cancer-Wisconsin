```python
from sklearn.preprocessing import LabelEncoder
datal = datad.apply(LabelEncoder().fit_transform)
datal.columns
```

### FEATURE SELECTION

```python
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE
```

```python
X = datal.values[:,1:30]
Y = datal.values[:,0]
```

```python
Y
```

#### PCA

```python
pca = PCA(n_components=5)
fit = pca.fit(X)
```

```python
print("Explained Variance:",fit.explained_variance_ratio_)
print("Components:",fit.components_)
```

#### RFE

```python
model = LogisticRegression()
rfe = RFE(model,9)
fit = rfe.fit(X,Y)
print("Features:",fit.n_features_)
print("Support:",fit.support_)
print("Ranking:",fit.ranking_)
```

### COMPARING DIFFERENT MODELS

```python
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler

from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
model = []
model.append(("GaussianNB",GaussianNB()))
model.append(("LogisticRegression",LogisticRegression()))
model.append(("SVM",SVC()))
model.append(("DecisionTreeClassifier",DecisionTreeClassifier()))
model.append(("LinearDiscriminantAnalysis",LinearDiscriminantAnalysis()))
model.append(('RandomForestClassifier',RandomForestClassifier()))
```

```python
results = []
names = []
for name,model in model:
    kfold = KFold(n_splits=10,random_state=7)
    cross = cross_val_score(model,X,Y,cv=kfold,scoring='accuracy')
    results.append(cross)
    names.append(name)
    print(name,":",cross.mean(),"(",cross.std(),')')
```
